```python
def count_datatypes(*args):
    dt={int:0, str:1, bool:2, list:3, tuple:4, dict:5}
    lst=[0,0,0,0,0,0]
    
    for i in args:
        typ=type(i)
        index=dt[typ]
        lst[index]+=1
    return lst
    
    
```

```python
count_datatypes(1, 45, "Hi", False)
```

```python
count_datatypes([10, 20], ("t", "Ok"), 2, 3, 1)
```

```python
def fib_str(n,strng):
    if n<=len(strng):
        return "enter number more than {} (number of strings)".format(len(strng))
    for i in range(n-2):
        a=''
        a=''.join((strng[-1],strng[-2]))
        strng.append(a)
    return strng
```

```python
fib_str(3, ["j", "h"])
```

```python
fib_str(5, ["e", "a"])
```

```python
fib_str(6, ["n", "k"])
```

```python
def ones_threes_nines(num):
    nines=0
    threes=0
    ones=0
    if 0<num<=26:
        rem9=num%9
        nines=num//9
        rem3=rem9%3
        threes=rem9//3
        ones=rem3//1
        return "nines:{}, threes:{}, ones:{}".format(nines,threes,ones)
    
    else:
        return "enter numbers between 0 to 26"
    
    
```

```python
def fib(num):
    count = 0
    f1 , f2 = 0,1
    if num <= 0:
        print("Number not greater than 0")
    elif num == 1:
        return num       
    else:
        while count <= num:   
            count+=1
            if count > num:
                return f1
            fth = f1 + f2
            f1,f2 = f2, fth 
```

```python
fib(6)
```

```python
fib(1)
```

```python
fib(2)
```
